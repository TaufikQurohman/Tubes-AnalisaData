**2.  Menghitung korelasi**

**I. Membuat rumus menghitung korelasi**

```python
def calculate_correlation(x, y):
   # Rumus korelasi: r(Xi,Xj) = [Σ(Xi×Xj) - (ΣXi×ΣXj)/n] / √[(ΣXi² - (ΣXi)²/n) × (ΣXj² - (ΣXj)²/n)]
   n = len(x)

   # Komponen dasar
   sum_xi = sum(x)
   sum_xj = sum(y)
   sum_xi_squared = sum([xi**2 for xi in x])
   sum_xj_squared = sum([xj**2 for xj in y])
   sum_xi_xj = sum([x[i] * y[i] for i in range(n)])

   # Pembilang: Σ(Xi×Xj) - (ΣXi×ΣXj)/n
   numerator = sum_xi_xj - (sum_xi * sum_xj) / n

   # Penyebut: √[(ΣXi² - (ΣXi)²/n) × (ΣXj² - (ΣXj)²/n)]
   denominator_xi = sum_xi_squared - (sum_xi**2) / n
   denominator_xj = sum_xj_squared - (sum_xj**2) / n
   denominator = math.sqrt(denominator_xi * denominator_xj)

   # Return korelasi atau 0 jika pembagi nol
   if denominator == 0:
       return 0
   else:
       return numerator / denominator
```

**3. Stepwise**

**I. Membuat rumus**

```python
# Buat rumus-rumus

# JKR sederhana
def Jumlah_Kuadrat_Regresi(col1, col2, n):
  x = np.sum(col1)
  y = np.sum(col2)
  x_bar = x / n
  y_bar = y / n
  xy = np.sum(col1 * col2)
  x2 = np.sum(col1 ** 2)

  b = (n * xy - x * y)/(n * x2 - x**2)
  a = y_bar - b * x_bar
  yi = []
  for i in range(n):
    yi.append(a + b * col1[i])
  hasil = np.sum((yi - y_bar)**2)
  return hasil

# JKR berganda
def Jumlah_Kuadrat_Regresi_Berganda(xs, y):
  xs = [np.array(x) for x in xs]
  X_mat = np.column_stack([np.ones(len(y))] + xs)
  beta = np.linalg.inv(X_mat.T @ X_mat) @ (X_mat.T @ y)
  Y_pred = X_mat @ beta
  JKR = np.sum((Y_pred - np.mean(y))**2)
  return JKR, beta

# JKT
def Jumlah_Kuadrat_Total(col1, n):
  y = np.sum(col1)
  y_bar = y / n
  hasil = 0
  for i in range(n):
    hasil =  (col1[i] - y_bar)**2 + hasil

  return hasil

JKT = Jumlah_Kuadrat_Total(kolom[-1], n)
print("nilai JKT =",f"{JKT:.4f}")
```

Nilai JKT yang besar (902.95 miliar) menunjukkan bahwa nilai gaji (Salary) memiliki keragaman yang tinggi (varian besar) dalam dataset ini

**II. Uji satu satu**

```python
# Uji satu satu
# Menghitung nilai R tiap peubah
JKR_sederhana = []
for i in kolom[:-1]:
  hasil = Jumlah_Kuadrat_Regresi(i, kolom[-1],n)
  JKR_sederhana.append(hasil)

# Melihat masing masing nilai R
print("")
print("TAHAP 1 - Uji peubah tunggal")
for i, JKR in enumerate(JKR_sederhana):
  print(f"R X{i+1} =" ,f"{JKR:.4f}")

# Mencari peubah yang memiliki nilai R terbesar
print("")
print("TAHAP 2 - cari R terbesar")
idx_terbesar = np.argmax(JKR_sederhana)
print(f"R terbesar adalah X{idx_terbesar+1} = {JKR_sederhana[idx_terbesar]:.4f}")

# Membuat array untuk menyimpan nilai nilai
variable_terpilih = []
variable_sisa = list(set(range(4)) - set(variable_terpilih))
JKR_sekarang = JKR_sederhana[idx_terbesar]

# Menghitung nilai F hitung dan F tabel
print("")
print(f"TAHAP 3 - hitung nilai F dari X{idx_terbesar+1}")
df = n - (len(variable_terpilih)+ 1 + 1)
S2 = (JKT - JKR_sekarang) / df
F_hitung = JKR_sekarang / S2
F_tabel = st.f.ppf(0.95,1,df)
print("R =",f"{JKR_sederhana[idx_terbesar]:.4f}")
print("S kuadrat =",f"{S2:.4f}")
print("nilai F hitung =",f"{F_hitung:.4f}")
print("nilai F tabel =", f"{F_tabel:.4f}")

# Keputusan apakah peubah masuk ke dalam model atau tidak
if F_hitung > F_tabel:
  print(f"peubah X{idx_terbesar+1} masuk dalam model")
  variable_terpilih.append(idx_terbesar)
else:
  print(f"peubah X{idx_terbesar+1} tidak masuk dalam model")
```

Dari uji peubah tunggal, tidak ada peubah yang secara signifikan memengaruhi Salary jika dilihat satu per satu.

Meski X1 memiliki kontribusi variasi terbesar, nilai F-nya tidak melewati ambang signifikansi, sehingga model regresi tidak menyertakan variabel apapun pada tahap ini.

```python
# Menghitung nilai F hitung dan F tabel dari peubah dengan R tertinggi ke 2
print("")
print("TAHAP 3 - hitung nilai F dari X2")
df = n - (len(variable_terpilih)+ 1 + 1)
S2 = (JKT - JKR_sederhana[1]) / df
F_hitung = JKR_sederhana[1] / S2
F_tabel = st.f.ppf(0.95,1,df)

print("R =",f"{JKR_sederhana[1]:.4f}")
print("S kuadrat =",f"{S2:.4f}")
print("nilai F hitung =",f"{F_hitung:.4f}")
print("nilai F tabel =", f"{F_tabel:.4f}")

if F_hitung > F_tabel:
  print(f"peubah X{1+1} masuk dalam model")
  variable_terpilih.append(1)
else:
  print(f"peubah X{1+1} tidak masuk dalam model")
```

**III. Uji Selanjutnya**

```python
idx_terbesar = np.argmax(JKR_sederhana)
selected_vars = [idx_terbesar]
remaining_vars = list(set(range(4)) - set(selected_vars))
JKR_terkini = JKR_sederhana[idx_terbesar]

df_error = n - (len(selected_vars) + 1)
S2 = (JKT - JKR_terkini) / df_error
Fh = JKR_terkini / S2
Ft = st.f.ppf(0.90, 1, df_error)
print(f"X{idx_terbesar+1} masuk model pertama: F hitung = {Fh:.4f} > F tabel = {Ft:.4f} \n")

# Tahap-tahap berikutnya
while len(selected_vars) < 4:
    best_1 = 0
    best_F = 0
    best_var = None
    best_JKR = None
    best_S2 = None
    for i in remaining_vars:
        vars_test = selected_vars + [i]
        JKR_new, _ = Jumlah_Kuadrat_Regresi_Berganda([kolom[j] for j in vars_test], kolom[-1])
        delta_JKR = JKR_new - JKR_terkini
        df_num = 1
        df_den = n - (len(vars_test) + 1)
        S2_new = (JKT - JKR_new) / df_den
        Fh = delta_JKR / S2_new
        Ft = st.f.ppf(0.90, df_num, df_den)
        print(f"Uji tambah X{i+1}: R = {delta_JKR:.4f}, F hitung = {Fh:.4f}, F tabel = {Ft:.4f}")
        if best_1 < delta_JKR:
          best_1 = delta_JKR
          if Fh > Ft and Fh > best_F:
              best_F = Fh
              best_var = i
              best_JKR = JKR_new
              best_S2 = S2_new

    if best_var is not None:
        selected_vars.append(best_var)
        remaining_vars.remove(best_var)
        JKR_terkini = best_JKR
        S2 = best_S2
        print(f"\n Peubah X{best_var+1} masuk model, F = {best_F:.4f}\n")
    else:
        print("\nTidak ada peubah signifikan lagi. Proses berhenti.\n")
        break
```

Model regresi terbaik yang terbentuk hanya menggunakan satu peubah, yaitu X1 (Graduation Year).

Peubah X2 (English), X3 (Logical), dan X4 (Computer Programming) tidak memberikan kontribusi signifikan tambahan terhadap prediksi gaji (Salary).
